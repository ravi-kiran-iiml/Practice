```python
param1 = ['complementing_partner','fintech_partner'] 
param2 = ['closed_market', 'open_market', 'whitelabel_market']
param3 = ['financial_products','non_financial_products']
param4 = ['partner_sell_only','partner_cocreate_and_sell']
param5 = ['bank_sell_only','bank_cocreate_and_sell']
param6 = ['bank_frontfacing', 'bank_embedded_in_partner']
param7 = ['large_corporate_partner','premium_brand_partner','other_companies_partner']

param_list = [param1, param2, param3,param4, param5, param6, param7]
```

```python
def list_expand(*args):
  global param_list
  factors_list = []
  for param in param_list:
    for factor in param:
      factors_list.append(factor)
  #print(factors_list)
  return factors_list

print(list_expand(param_list))
```

```python
def factor_combination(*args):
  global param_list
  factors_list = list_expand(param_list)
  dimensions = len(param_list)
  n_models = 1
  for param in param_list:
    n_models = n_models * len(param) 
  #print("Factors: ", factors_list)
  print("Dimensions: ",dimensions, "," ,"No of Models: ",n_models)
  combinations = []
  for i in range(n_models):
    one_model = []
    for j in range(dimensions):
      one_model.append(factors_list[j])
    combinations.append(one_model)
  #print(combinations)
  return combinations

print(factor_combination())
```

```python
def eliminate_models(*args):
  global param_list
  combinations = factor_combination(param_list)
  for combination in combinations:
    for param in param_list:
      pass
  return combinations

eliminate_models()
```

```python
import pandas as pd

def output_models(*args):
  global param_list
  models = eliminate_models()
  model_df = pd.DataFrame(models)
  #print(model_df)
  #model_df.to_csv('Models.csv')
  return model_df

output_models()
```

```python
def upload_score_matrix(*args):
  score_matrix = pd.read_csv('Parameter Scores.csv')
  #print(score_matrix)
  return score_matrix

upload_score_matrix()
```

```python
def factor_synergy_dict(*args):
  df = upload_score_matrix()
  df_factor = df.T
  scores_dict = {}
  index_list = list(df_factor.index.values)
  index_list = index_list[1:]
  #print(index_list)
  df_factor.columns = df_factor.iloc[0]
  df_factor = df_factor.drop(df_factor.index[0])
  factors_list = list(df_factor.columns)
  #print(factors_list)

  factor_synergy_list = []

  for factor in factors_list:
    z = zip(index_list, df_factor[factor])
    #factor_synergy_list.append([factor,[]])
    for item in list(z):
      factor_synergy_list.append([factor, item])
        
  return df_factor #factor_synergy_list,  

factor_synergy_dict()
```

```python

def calc_model_score(*args):
  global param_list
  df_factor = factor_synergy_dict()
  models_dataframe = output_models()
  f = upload_score_matrix()
  #print(models_dataframe)

  synergies_list = list(f.columns)[1:]
  #print(synergies_list)

  import numpy as np

  d = pd.DataFrame(0, index=np.arange(len(models_dataframe.index)), columns=synergies_list)
  #print(d)
  models_dataframe = pd.concat([models_dataframe, d], axis=1)
  #print(models_dataframe)
  #print(models_dataframe.columns)
  #print(df_factor)


  # Score Addition to DataFrame
  for i in range(len(models_dataframe)):
    for synergy in synergies_list:
      models_dataframe.at[i, synergy] = 0
      synergy_score = 0
      for j in range(len(param_list)):
        synergy_score += df_factor.at[synergy, models_dataframe.at[i,j]]
      models_dataframe.at[i,synergy] = synergy_score
      pass
  #scored_model = pd.DataFrame()
  return models_dataframe #scored_model

calc_model_score()
```
